```python
def amount_creator(d):
    price = d['price']
    quantity = d['quantity']
    unit = d['unit']
    if unit == 'gm':
        quantity /= 1000
    amount = price  * quantity
    return amount
```

```python
data['amount'] = data.apply(amount_creator, axis =1)
data.head()
```

```python
data['name'][data[data['category']=='chaat']['amount'].idxmax()]
```
